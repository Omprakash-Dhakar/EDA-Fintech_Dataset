##  Univariate Analysis

**Function to plot the univariate categorical variables**

```python
# Defining a function 
def Univariate_Graph(var):

    # Setting style for the plot using seaborn
    sns.set_style('darkgrid')
    
    # Creating a figure and a pair of subplots and specifying the desired figure size
    fig, (axes1,axes2) = plt.subplots(1, 2, figsize=(18,6))
    
    # Plotting a count of each unique value in column for Target0_df
    sns.countplot(x = var, data = Target0_df, ax = axes1, palette = 'rainbow')
    axes1.set_ylabel('Total Counts')
    axes1.set_title(f'Distribution of {var} for Non-Defaulters', fontsize = 16)
    axes1.set_xticklabels(axes1.get_xticklabels(), rotation = 90)
    
    # Adding annotations with the normalized percentage of each patch
    for patch in axes1.patches:
        axes1.annotate('{:.2f}%'.format((patch.get_height() / len(Target0_df)) * 100), 
                                        (patch.get_x() + 0.1, patch.get_height() + 50))
     
    # Plotting a count of each unique value in column for Target1_df
    sns.countplot(x  =var, data = Target1_df, ax = axes2, palette = 'rainbow')
    axes2.set_ylabel('Total Counts')
    axes2.set_title(f'Distribution of {var} for Defaulters', fontsize = 16) 
    axes2.set_xticklabels(axes2.get_xticklabels(), rotation = 90)
    
    # Adding annotations with the normalized percentage of each patch
    for patch in axes2.patches:
        axes2.annotate('{:.2f}%'.format((patch.get_height() / len(Target1_df)) * 100), 
                                        (patch.get_x() + 0.1, patch.get_height() + 50))
```

##  Univariate  Analysis For Continuous Variable

```python
def Plot_Univariate_Distribution(var, style='darkgrid', figsize=(18, 6), title_fontsize=14):    
    # set plot style
    sns.set_style(style)
    
    # creating figure, subplots and setting figure size
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    
    # plot distribution for non-defaulters
    sns.distplot(a=Target0_df[var], color='blue', ax=ax1)
    ax1.set_title(f'Distribution of {var} for Non-Defaulters', fontsize=title_fontsize)
    
    # plot distribution for defaulters
    sns.distplot(a=Target1_df[var], ax=ax2)
    ax2.set_title(f'Distribution of {var} for Defaulters', fontsize=title_fontsize)
    
    plt.show()
```

###  Getting the top 10 correlation of the selected columns

```python
# Defining a function
def Top_Correlations(data, n=10):
    
    # calculating correlations
    corr = data.corr()
    
    # selecting upper triangle of correlation matrix
    corr_df = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool)).unstack().reset_index()
    
    # rename columns
    corr_df.columns = ['Column_1', 'Column_2', 'Correlation']
    
    # drop rows with missing values
    corr_df.dropna(subset = ['Correlation'], inplace=True)
    
    # adding new column name as absolute correlation 
    corr_df['Abs_Corr'] = corr_df['Correlation'].abs()
    
    # sorting by absolute correlation
    corr_df = corr_df.sort_values(by=['Abs_Corr'], ascending=False)
    
    return corr_df.head(n)
```

##  Bivariate Analysis 

```python
# function for scatter plot for continuous variables
def Plot_Bivariate_Scatter(var1,var2):

    # setting style for plot
    sns.set_style('darkgrid')
    
    # Creating a figure and a pair of subplots and specifying the desired figure size
    fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,6))
    
    # plot scatter plot for Non-Defaulters
    sns.scatterplot(x = var1, y = var2, data = Target0_df, color = 'blue', ax = ax1)
    ax1.set_xlabel(var1)    
    ax1.set_ylabel(var2)
    
    #setting the title
    ax1.set_title(f'{var1} vs {var2} for Non-Defaulters', fontsize = 14)
    
    #Plot scatter for Defaulters 
    sns.scatterplot(x = var1, y = var2, data = Target1_df, color = 'red', ax = ax2)
    ax2.set_xlabel(var1)    
    ax2.set_ylabel(var2)
    
    # setting the title
    ax2.set_title(f'{var1} vs {var2} for Defaulters', fontsize = 14)
```

```python
def Univariate_Merged(var1, var2):
    
    sns.set_style('darkgrid')

    New_df = Merged_df.pivot_table(values = 'SK_ID_CURR', 
                      index=var1,
                      columns=var2,
                      aggfunc='count')
    
    New_df = New_df.div(New_df.sum(axis = 1), axis = 'rows') * 100
    sns.set()
    
    New_df.plot(kind='bar', stacked=True, figsize=(18,6), cmap='Set1')
    plt.title(f'Effect Of {var1} on Loan Approval')
    plt.xlabel(f'{var1}')
    plt.ylabel(f'{var2}%')
    plt.show()
```

```python
Univariate_Merged('FLAG_OWN_CAR','NAME_CONTRACT_STATUS')
```

**Analysis of default rates shows that borrowers who own a car have lower default rates compared to those who do not own a car. As a result, the bank may want to consider giving additional weight to car ownership when approving loan amounts, in order to potentially reduce the risk of default.**

```python
Univariate_Merged('CODE_GENDER','NAME_CONTRACT_STATUS')
```

While gender does not seem to have a significant effect on loan application approval or rejection, analysis of default rates shows that female borrowers have lower default rates compared to male borrowers. Therefore, the bank may want to consider giving additional weight to gender when approving loan amounts, in order to potentially reduce the risk of default
